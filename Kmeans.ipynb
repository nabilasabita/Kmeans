```
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plot
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.cluster import KMeans
import seaborn as sns
```

```
x = df[36]
y = df[40]
```

```
df_df = np.c_[x,y]
df_df
```

```
sns.scatterplot(x=df_df[:,0],y=df_df[:,1],s=100,color="blue",alpha=0.5)
```

```
df_nor=((df_df-df_df.min())/(df_df.max()-df_df.min()))
df_df
```

```
%matplotlib inline 
from sklearn.cluster import KMeans
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_df)
    distortions.append(kmeanModel.inertia_)
```

```
plot.figure(figsize=(16,8))
plot.plot(K, distortions, 'bx-')
plot.xlabel('k')
plot.ylabel('Distortion')
plot.title('The Elbow Method showing the optimal k')
plot.show()
```

```
# Fitting K-Means to dataset
kmeans = KMeans(n_clusters= 4, init= 'k-means++', random_state= 52)
y_kmeans = kmeans.fit_predict(df_df)
```

```
#Visualiting The Clusters
plot.scatter(df_df[y_kmeans == 0, 0], df_df[y_kmeans == 0, 1], s = 10, c = 'red', label = 'Cluster 1')
plot.scatter(df_df[y_kmeans == 1, 0], df_df[y_kmeans == 1, 1], s = 10, c = 'blue', label = 'Cluster 2')
plot.scatter(df_df[y_kmeans == 2, 0], df_df[y_kmeans == 2, 1], s = 10, c = 'green', label = 'Cluster 3')
plot.scatter(df_df[y_kmeans == 3, 0], df_df[y_kmeans == 3, 1], s = 10, c = 'cyan', label = 'Cluster 4')
plot.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 30, c = 'black', label = 'Centroids')
plot.title('Clusters')
plot.xlabel('K.FORMA_FINAL')
plot.ylabel('N.SUPERFICIE')
plot.legend()
plot.show()
print("list cluster center")
print(kmeans.cluster_centers_)
print("cluster labels")
print(kmeans.labels_)
```

```
# Calculate silhouette_score
from sklearn.metrics import silhouette_score

print(silhouette_score(df_df, kmeans.labels_))
```
